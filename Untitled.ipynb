```python
menu = {
    "Pizza": 200,
    "Burger": 150,
    "Pasta": 180,
    "Coffee": 80
}

#  BILL CALCULATOR TOOL
def restaurant_bill(input_text):
    # Input format: Item,Quantity
    item, qty = input_text.split(",")
    item = item.strip()
    qty = int(qty.strip())

    if item in menu:
        subtotal = menu[item] * qty
        gst = subtotal * 0.05
        total = subtotal + gst

        return f"""
Item: {item}
Quantity: {qty}
Subtotal: ₹{subtotal}
GST (5%): ₹{gst}
Total Bill: ₹{total}
"""
    else:
        return "Item not available in menu."

restaurant_tool = Tool(
    name="Restaurant Bill Calculator",
    func=restaurant_bill,
    description="Calculate restaurant bill. Input format: Item,Quantity"
)

# AGENT Tool
agent = initialize_agent(
    tools=[restaurant_tool, BILL CALCULATOR TOOL, AGENT Tool],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=False
)
```
